Below is the dating person data which includes 2 genders

```python
df = pd.read_csv("darting_persons.csv")
df.head()
```

```python
plt.scatter(df.goal,df.age)
plt.xlabel('goal')
plt.ylabel('Age')
```

```python
df = df.drop('iid', 1)
df.head()
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df)
y_predicted
```

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['goal','age']])
    sse.append(km.inertia_)
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```
